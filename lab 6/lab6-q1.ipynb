```python
x = df['experience'].values
y = df['salary'].values
mean_x = np.mean(x)
mean_y = np.mean(y)
b1 = np.sum((x-mean_x)*(y-mean_y))/(np.sum((x-mean_x)**2))
b0 = mean_y -b1*mean_x
y_pred = b0 + b1*x
error = y_pred-y
mse = np.mean((error)**2)
rmse=np.sqrt(mse/len(x))
print(f"b0: {b0} b1: {b1}")
predicted_df = pd.DataFrame({
    'experience' : x,
    'salary' : y,
    'predicted_salary' : y_pred
})
print("Predicted Response :")
print(predicted_df)
```

```python
print(f"RMSE: {rmse}")
```
