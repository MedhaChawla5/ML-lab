```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
df = pd.read_csv('data1.csv')
print(df)
```

```python
x = df['experience'].values
y = df['salary'].values
y_min = min(y)
y_max = max(y)
y = y-y_min/(y_max-y_min)
b0 = 0
b1 = 0
lr = 0.01
y_pred = []
log_loss_values =[]
for j in range(5):
    print(f"Epoch {j+1}:")
    for i in range(len(x)):
        z = b0+b1*x[i]
        pred = 1/(1 + np.exp(-z))
        err = pred - y[i]
        b0 = b0 - lr*err
        b1 = b1 -lr*err*x[i]
        print(f"b0: {b0} b1: {b1}")
        log_loss = y[i] * np.log(pred) + (1 - y[i]) * np.log(1 - pred)
        log_loss_values.append(log_loss/err)
print(log_loss_values)
```

```python
array = np.arange(1,61,1)
plt.plot(array,log_loss_values,color = 'red')
plt.ylabel("log loss versus error")
plt.xlabel("iterations")
plt.show()
```
