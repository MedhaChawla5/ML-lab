```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
x = [0,20,40,60,80,100]
y = [i*2 for i in x]
ax.plot(x,y)
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("question 1")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
x = [0,20,40,60,80,100]
y = [i*2 for i in x]

ax1.plot(x,y)
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('Main Plot')

ax2.plot(x,y)
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_title('Inset Plot')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

x = df['month_number']
y = df['total_units']

fig, ax = plt.subplots()
ax.plot(x, y, linestyle=':',color='red',marker='o',markerfacecolor='red', linewidth=3,label='Total Profit') 

ax.set_xlabel('Month Number')  
ax.set_ylabel('Sold Units Number') 
ax.set_title('Question 4')

ax.legend(loc='lower right')

plt.show()

year_profit = df['total_profit'].sum()
print(f"The total profit in the given year is: {year_profit}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")


months = df['month_number']
products = df.columns[1:7]  

fig, ax = plt.subplots()

for product in products:
    ax.plot(df['month_number'], df[product], label=product)

ax.set_xlabel('month number')
ax.set_ylabel('Number of Units Sold')
ax.set_title('Units Sold Per Month for Each Product')

ax.legend(loc='upper left')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

df.drop(columns=['month_number','total_units','total_profit'], inplace=True)
total_sales_last_year = df.sum(numeric_only=True)  


fig, ax = plt.subplots()
ax.pie(total_sales_last_year, 
       labels=total_sales_last_year.index, 
       autopct='%1.1f%%') 


ax.set_title('Percentage of Units Sold per Product for Last Year')
plt.show()
```
